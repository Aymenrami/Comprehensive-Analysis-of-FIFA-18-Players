```python
average_age_per_club = df.groupby('Club')['Age'].mean().sort_values(ascending=False)
top_clubs_by_player_count = df['Club'].value_counts().head(20).index.sort_values(ascending = False)
filtered_club_ages = average_age_per_club.loc[top_clubs_by_player_count].reset_index()
filtered_club_ages
```

```python
plt.figure(figsize=(6, 5))
sns.barplot(data= filtered_club_ages, x='Age', y='Club', palette='viridis' , hue = 'Club')
plt.xticks(rotation=90)
plt.title('Average Age of Players in Top 20 Clubs by Player Count')
plt.xlabel('Club')
plt.ylabel('Average Age')
plt.show()
```
